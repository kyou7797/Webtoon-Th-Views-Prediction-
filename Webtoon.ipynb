```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('webtoon_originals_th.csv')
```

```python
print(df.shape)
df.head()
```

```python
df = df.drop(['synopsis'], axis = 1)
df.head()
```

```python
df['daily_pass'] = df['daily_pass'].apply(lambda val: 1 if val == 1 else 0)
df.head()
```

```python
df.info()
```

```python
df = df.drop(['title','authors','title_id'],axis=1)
```

```python
df.head()
```

```python
df['status'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder 
le_status = LabelEncoder()
print(df['status'].unique())
df['status'] = le_status.fit_transform(df['status'])
df['status'].unique()
```

```python
# data with weekdays
df_day = df
```

```python
df = df.drop(['weekdays'],axis = 1)
```

```python
df.genre.value_counts()
```

```python
le_genre = LabelEncoder()
print(df['genre'].unique())
df['genre'] = le_genre.fit_transform(df['genre'])
df['genre'].unique()
```

```python
df.head()
```

```python
df.describe()
```

```python
gp = df.describe()
gp = gp.loc[['25%','75%']]
gp = gp.T
gp['IQR'] = gp['75%'] - gp['25%']
gp['Min'] = gp['25%'] - 1.5 * gp['IQR']
gp['Max'] = gp['75%'] + 1.5 * gp['IQR']

listNon = {'genre','daily_pass','status'}

for col in df.columns:
    if col in listNon:
        continue
    df = df[(df[col] >= gp['Min'][col]) & (df[col] <= gp['Max'][col])]
df.shape
```

```python
df.describe()
```

```python
X = df.drop('views', axis=1)
y = df['views']
```

```python
# LinearRegression

from sklearn.linear_model import LinearRegression
linear_reg = LinearRegression()
linear_reg.fit(X, y.values)

y_pred = linear_reg.predict(X)

from sklearn.metrics import mean_squared_error, mean_absolute_error
error = np.sqrt(mean_squared_error(y, y_pred))
print("{:,.02f} views".format(error))
```

```python
# DecisionTree

from sklearn.tree import DecisionTreeRegressor
dec_tree_reg = DecisionTreeRegressor(random_state=0)
dec_tree_reg.fit(X, y.values)

y_pred = dec_tree_reg.predict(X)

error = np.sqrt(mean_squared_error(y, y_pred))
print("{:,.02f} views".format(error))
```

```python
# RandomForestRegressor

from sklearn.ensemble import RandomForestRegressor
random_forest_reg = RandomForestRegressor(random_state=0)
random_forest_reg.fit(X, y.values)

y_pred = random_forest_reg.predict(X)

error = np.sqrt(mean_squared_error(y, y_pred))
print("{:,.02f} views".format(error))
```

```python
# Decision Tree GridSearch

from sklearn.model_selection import GridSearchCV

max_depth = [None, 2,4,6,8,10,12]
parameters = {"max_depth": max_depth}

regressor = DecisionTreeRegressor(random_state=0)
gs = GridSearchCV(regressor, parameters, scoring='neg_mean_squared_error')
gs.fit(X, y.values)

regressor = gs.best_estimator_

regressor.fit(X, y.values)
y_pred = regressor.predict(X)
error = np.sqrt(mean_squared_error(y, y_pred))
print("{:,.02f} views".format(error))
```
